```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('all_data.csv')
print(data.head())
```

```python
#here I created six new variables. The goal is to 
#have one varable represent each country, repectively. 
#Going further I will do bar charts for how each country's life expectancy and gdp 
#changed over the years and if they were related or not. 

chile_data = data[data['Country'] == 'Chile']
china_data = data[data['Country'] == 'China']
germany_data = data[data['Country'] == 'Germany']
mexico_data = data[data['Country'] == 'Mexico']
USA_data = data[data['Country'] == 'United States of America']
zimbabwe_data = data[data['Country'] == 'Zimbabwe']




#here I did a barplat with sns to see the average Life Expectancy in each country 
#over 15 years. 

sns.barplot(data = data, x='Country', y='Life expectancy at birth (years)', capsize=0.5)
plt.xticks(rotation=290)
plt.ylim(40,84)
plt.title('Average Life Expectancy in each Country')
plt.legend(['Countries','Life Expectancy'])
plt.ylim(40,84)
plt.show()

plt.clf()

```

```python
#calculating covarian of those coutnries whose bar plots show a little less consistency bbetween gdp growth and life expectancy growth. 
import numpy as np
cov_chile_year_gdp = np.cov(chile_data.Year, chile_data['Life expectancy at birth (years)'])
print(str(cov_chile_year_gdp) + "is the covariance of Chile")

# germany 
cov_germany_year_gdp = np.cov(germany_data.Year, germany_data['Life expectancy at birth (years)'])
print(str(cov_germany_year_gdp) + "is the covariance of Germany")

#zimbabwe is the lest linear looks like:
cov_zimbabwe_year_gdp = np.cov(zimbabwe_data.Year, zimbabwe_data['Life expectancy at birth (years)'])
print(str(cov_zimbabwe_year_gdp) + "is the covariance of Zimbabwe")


#lets see how strong these covariances really are.


from scipy.stats import pearsonr
corr_chile_year_gdp, p = pearsonr(chile_data.Year, chile_data['Life expectancy at birth (years)'])
print(str(corr_chile_year_gdp)+" is the correlation of chile's years and life expectancy")
      
corr_germany_year_gdp, p = pearsonr(germany_data.Year, germany_data['Life expectancy at birth (years)'])
print(str(corr_germany_year_gdp)+" is the correlation of germany's years and life expectancy")

corr_zimbabwe_year_gdp, p = pearsonr(zimbabwe_data.Year, zimbabwe_data['Life expectancy at birth (years)'])
print(str(corr_zimbabwe_year_gdp)+" is the correlation of zimbabwe's years and life expectancy")


```
